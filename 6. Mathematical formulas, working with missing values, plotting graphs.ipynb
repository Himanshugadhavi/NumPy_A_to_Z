# 1. Working with mathematical formulas

## a. sigmoid
* A sigmoid function is a mathematical function with a characteristic "S"-shaped curve or sigmoid curve. It transforms any value in the domain ( -infinity, +infinity) to a number between 0 and 1.
![](sigmoid.png)

```python
def sigmoid(array):
    return 1/(1+np.exp(-(array)))
```

```python
a = np.arange(10)
a
```

```python
sigmoid(a)
```

## b. mean squared error
* In statistics, the concept of mean squared error is an essential measure utilized to determine the performance of an estimator. It is abbreviated as MSE and is necessary for relaying the concepts of precision, bias and accuracy during the statistical estimation.
![](mse.svg)

```python
actual = np.random.randint(1,50,25)
predicted = np.random.randint(1,50,25)

print(actual)
print(predicted)
```

```python
def mse(actual,predicted):
    return np.mean((actual - predicted)**2)
```

```python
mse(actual,predicted)
```

# 2. working with missing values

```python
# we are creating an array with missing value
a = np.array([1,2,3,4,np.nan,6])
a
# way to create missing value = np.nan
```

```python
np.isnan(a)
# by this we will get a Boolean array
```

```python
a[np.isnan(a)]
# this will give us value which are nan
```

```python
# if we use ~ (not) then it will return the non nan values
a[ ~np.isnan(a)]
```

>* seed() takes at most 1 positional argument
>* first time when we generate a random number using a random.seed it will generate any random number and it will save it in it's own memory 
>* now again whenever we run that code it will give the saved random value every time
